```python
class ListNode:
    
    def __init__(self, key, val):
        self.key = key
        self.val = val
        self.next = None
        self.prev = None

        
        
class LinkedList:
    
    def __init__(self):
        self.head = None
        self.tail = None
        
        
    def insert(self, node):
        
        if self.head is None:
            self.head = node
        else:
            
            
            self.tail.next = node
            node.prev = self.tail
            
        self.tail = node
        
    def delete(self, node):
        
        
        if node.prev:
            node.prev.next = node.next
        else:
            self.head = node.next
            
        if node.next:
            node.next.prev = node.prev
        else:
            self.tail=node.prev
            
    
class LRUCache(object):

    def __init__(self, capacity):
        """
        :type capacity: int
        """
        self.cacheDict ={}
        self.list = LinkedList()
        self.capacity = capacity
        
    def _insert(self,key,val):
        node = ListNode(key,val)
        self.list.insert(node)
        self.cacheDict[key]=node
        
        

    def get(self, key):
        """
        :type key: int
        :rtype: int
        """
        if key in self.cacheDict:
            val = self.cacheDict[key].val
            
            self.list.delete(self.cacheDict[key])
            self._insert(key,val)
        
            return val
        return -1 

    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """
        if key in self.cacheDict:
            self.list.delete(self.cacheDict[key])
            
            #self.list._insert(key,value)
        elif len(self.cacheDict)==self.capacity:
            del self.cacheDict[self.list.head.key]
            self.list.delete(self.list.head)
            
            
        self._insert(key,value)
            
        
        
```

```python
class Solution(object):
    def largestRectangleArea(self, heights):
        """
        :type heights: List[int]
        :rtype: int
        """
         
        #determione Last smaller element(LSE) 
        n = len(heights)
        LSE = [0]*n
        NSE = [n-1]*n
        #store in temp stack
        temp = []
        
        for i in range(n):
            
            while len(temp)>0 and heights[temp[len(temp)-1]]>=heights[i]:
                
                temp.pop()
                
            if len(temp)>0:
                LSE[i] = temp[len(temp)-1]+1
            else:
                LSE[i] = 0
            
            temp.append(i)
            
            #print(LSE)
            
            
        #look for NSE
        
        temp = []
        
        for i in range(n-1,-1,-1):
            while len(temp)>0 and heights[temp[len(temp)-1]]>=heights[i]:
                temp.pop()
                
                
            if (len(temp)>0):
                NSE[i] = temp[len(temp)-1] - 1
            else:
                NSE[i] = n - 1
                
            temp.append(i)
            
            #print(NSE)
        ans=0    
        for i in range(n):
            ans = max(ans, (NSE[i]-LSE[i]+1)*heights[i])
            
            
        print(ans)
        
        return ans
            
```

```python
class MinStack(object):

    def __init__(self):
        self.stk = []
        self.stk_desc = []
        

    def push(self, val):
        """
        :type val: int
        :rtype: None
        """
        self.stk.append(val)
        
        if len(self.stk_desc)==0 or val <= self.stk_desc[-1]:
            self.stk_desc.append(val)
        

    def pop(self):
        """
        :rtype: None
        """
        x = self.stk.pop()
        
        if x == self.stk_desc[-1]:
            self.stk_desc.pop()
        

    def top(self):
        """
        :rtype: int
        """
        return self.stk[-1]

    def getMin(self):
        """
        :rtype: int
        """
        
        return self.stk_desc[-1]
       
        
```

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        
        subj = " ".join(s.split()).split()
        ans =""
        for i in range(len(subj)-1,-1,-1):
            ans += subj[i]
            if    i > 0 and i < len(subj) :
                ans+=" "
            
            
        return ans
            
```

```python
class Solution(object):
    def partition(self, s):
        """
        :type s: str
        :rtype: List[List[str]]
        """
        
        
        #res, path  = [],[]
        
        res = []
        
        self.func(0, s, [], res)
        
        return res
    
    def isPalindrome(self,s,index,i):
        if s[index:i] == s[index:i][::-1]:
            return True
        return False
        #or return s == s[::-1]
        
        
    def func(self,index,s , path, res):
        
        if index == len(s):
            #print(index)
            res.append(list(path))
            #print(res)
            return
        
        
        
        for i in range(index,len(s)+1):
            #print(index,i)
            
            if self.isPalindrome(s,index,i+1):
                
                path.append(s[index:i+1])
                self.func(i+1,s,path,res)
                path.pop()
                
                
```
